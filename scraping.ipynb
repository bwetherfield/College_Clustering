```python
def get_names_term(n, major, session):
    url = "https://www.niche.com/colleges/search/{}/?page={}".format(major, n)
    response = session.get(url)
    html = response.html
    list_items = html.find('li.search-results__list__item')
    time.sleep(2)
    return [item.find('h2')[0].text for item in list_items]
```

```python
def get_num(n, term, session):
    return reduce(operator.add, [get_names_term(k, term, session) for k in list(range(1, n+1))])
```
